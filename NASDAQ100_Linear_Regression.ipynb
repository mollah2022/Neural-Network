```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/nasdaq100.csv",sep=";")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.drop( columns=[ "Date"])
```

```python
df.head()
```

```python
plt.scatter(df["Starting (USD)"],df['Ending (USD)'])
plt.xlabel("Starting (USD)")
plt.ylabel("Ending (USD)")
plt.title("NASDAQ100 Stock Prices")
plt.show()
```

```python
x = df[["Starting (USD)"]]
y = df[["Ending (USD)"]]
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
x.mean()
y.mean()
```

```python
plt.scatter(x.mean(),y.mean(),color="red")
plt.scatter(df["Starting (USD)"],df['Ending (USD)'])
plt.xlabel("Starting (USD)")
plt.ylabel("Ending (USD)")
plt.title("NASDAQ100 Stock Prices")
plt.show()
```

```python
reg.fit(x,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[16700]])
```

```python
df["Predict_y"] = reg.predict(x)
df.head()
```

```python
plt.plot(x,df["Predict_y"])
plt.scatter(x.mean(),y.mean(),color="red")
plt.scatter(df["Starting (USD)"],df['Ending (USD)'])
plt.xlabel("Starting (USD)")
plt.ylabel("Ending (USD)")
plt.title("NASDAQ100 Stock Prices")
plt.show()
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
```

```python
mse = mean_squared_error(df['Ending (USD)'],df['Predict_y'])
mse
```

```python
mae = mean_absolute_error(df['Ending (USD)'],df['Predict_y'])
mae
```

```python
reg.score(x,y)
```

```python
from sklearn.metrics import r2_score
r2_score(df['Ending (USD)'],df['Predict_y'])
```
