```python
# 1. দরকারি লাইব্রেরি ইমপোর্ট করি
import tensorflow as tf
from tensorflow.keras import layers, models
import matplotlib.pyplot as plt

# 2. ডেটাসেট লোড করি (MNIST handwritten digits)
(X_train, y_train), (X_test, y_test) = tf.keras.datasets.mnist.load_data()

# 3. ডেটাসেট Reshape এবং Normalize করা
X_train = X_train.reshape(-1, 28, 28, 1).astype('float32') / 255.0
X_test = X_test.reshape(-1, 28, 28, 1).astype('float32') / 255.0

# 4. মডেল তৈরি করা (CNN)
model = models.Sequential([
    layers.Conv2D(32, (3,3), activation='relu', input_shape=(28,28,1)), # Convolutional Layer 1
    layers.MaxPooling2D((2,2)),                                          # MaxPooling Layer 1
    layers.Conv2D(64, (3,3), activation='relu'),                         # Convolutional Layer 2
    layers.MaxPooling2D((2,2)),                                          # MaxPooling Layer 2
    layers.Flatten(),                                                    # Flatten করে 1D বানানো
    layers.Dense(64, activation='relu'),                                 # Fully Connected Layer (Dense)
    layers.Dense(10, activation='softmax')                               # Output Layer (10 Class)
])

# 5. মডেল কম্পাইল করা
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# 6. মডেল ট্রেন করা
history = model.fit(X_train, y_train, epochs=10, validation_data=(X_test, y_test))

# 7. Accuracy এবং Loss প্লট করা
plt.figure(figsize=(12,5))

# Accuracy Plot
plt.subplot(1,2,1)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Accuracy Curve')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)

# Loss Plot
plt.subplot(1,2,2)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Loss Curve')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

plt.show()

# 8. Model Summary দেখা
model.summary()
```
