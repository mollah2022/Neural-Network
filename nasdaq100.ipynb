```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df =pd.read_csv("/content/drive/MyDrive/Linear_regression/Linear-Regression-with-Python-main/nasdaq100.csv",sep=";")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.drop(columns=["Date"])
```

```python
df.head()
```

```python
plt.scatter(df["Starting (USD)"],df["Ending (USD)"])
plt.title("Nasdaq100")
plt.xlabel("Starting (USD)")
plt.ylabel("Ending (USD)")
plt.show()
```

```python
x = df[["Starting (USD)"]]
print(x)
y = df[["Ending (USD)"]]
print(y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
x.mean()
y.mean()
plt.scatter(x.mean(),y.mean(),color="red")
plt.scatter(df["Starting (USD)"],df["Ending (USD)"])
plt.title("Nasdaq100")
plt.xlabel("Starting (USD)")
plt.ylabel("Ending (USD)")
plt.show()
```

```python
reg.fit(x,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
df["Predict_y"] = reg.predict(x)
```

```python
df.head()
```

```python
plt.plot(x,df["Predict_y"],color="green")
plt.scatter(x.mean(),y.mean(),color="red")
plt.scatter(df["Starting (USD)"],df["Ending (USD)"])
plt.title("Nasdaq100")
plt.xlabel("Starting (USD)")
plt.ylabel("Ending (USD)")
plt.show()
```

```python
df["Loss"] = abs(df["Predict_y"] - df["Ending (USD)"])
df.head()
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

```python
mse = mean_squared_error(df["Ending (USD)"],df["Predict_y"])
print(mse)
mae = mean_absolute_error(df["Ending (USD)"],df["Predict_y"])
print(mae)
```
