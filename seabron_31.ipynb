```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seabron as sns
```

```python
df = sns.load_dataset("tips")
```

```python
df.head()
```

```python
sns.relplot(data=df,x="total_bill",y="tip")
```

```python
sns.relplot(data=df,x="total_bill",y="tip",hue="smoker")
```

```python
sns.relplot(data=df,x="total_bill",y="tip",hue="smoker",style="smoker")
```

```python
df = sns.load_dataset("dowjones")
df.head()
sns.relplot(data=df, x="Date", y="Price", kind="line")
```

```python
df = sns.load_dataset("fmri")
df.head()
```

```python
sns.relplot(data=df, x="timepoint", y="signal", kind="line")
```

```python
sns.relplot(
    data=df, kind="line",
    x="timepoint", y="signal",
    estimator=None,
)
```

```python
df.head()
```

```python
sns.relplot(
    data=df, kind="line",
    x="timepoint", y="signal", hue="event",
)
```

```python
df.head()
```

```python
sns.relplot(
    data=df, kind="line",
    x="timepoint", y="signal",
    hue="region", style="event",
)
```

```python
sns.relplot(
    data=df, kind="line",
    x="timepoint", y="signal", hue="region", style="event",
     markers=True,
)
```

```python
sns.relplot(
    data=df, kind="line",
    x="timepoint", y="signal", hue="region", style="event",
    dashes=False, markers=True,
)
```

```python
sns.relplot(
    data=df.query("event == 'stim'"), kind="line",
    x="timepoint", y="signal", hue="region",
    units="subject", estimator=None,
)
```

```python
dots = sns.load_dataset("dots").query("align == 'dots'")
sns.relplot(
    data=dots, kind="line",
    x="time", y="firing_rate",
    hue="coherence", style="choice",
)
```

```python
dots.head()
```

```python
healthexp = sns.load_dataset("healthexp").sort_values("Year")
sns.relplot(
    data=healthexp, kind="line",
    x="Spending_USD", y="Life_Expectancy", hue="Country",
    sort=False
)
```
